# Pipeline (Scaling+Feature Engineering)

## Scaling

```python
num_features=X.select_dtypes(include='float64').columns.tolist()
```

```python
num_features
```

```python
corr_num = X[num_features].corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr_num, cmap='coolwarm',vmin=-1, vmax=1);
```

```python
cat_features=X.select_dtypes(include='object').columns.tolist()
```

```python
cat_features
```

```python
#Custom Function for engineering the 'Cabin' feature
def cabin_transf(df):
    df['side']=df['Cabin'].str[-1]
    df['deck']=df['Cabin'].str[0]
    df.drop(columns='Cabin',axis=1,inplace=True)
    return df    
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler

# Categorical Custom Features
cabin_transformer = FunctionTransformer(lambda df:cabin_transf(df))
indexer = FunctionTransformer(lambda df: df.set_index('PassengerId'))
name_dropper = FunctionTransformer(lambda df: df.drop(columns=['Name']))
                                   
#Categorical Features Transformation
cat_transformer = Pipeline([('cabin_transformer',cabin_transformer),
                            ('indexer',indexer),
                            ('name_dropper',name_dropper),
                            ('imputer', SimpleImputer(strategy='most_frequent')),
                            ('scaler', OneHotEncoder(handle_unknown = 'ignore'))])

#Numerical Features Transformation
num_transformer= Pipeline([('imputer', SimpleImputer(strategy='mean')),('scaler', MinMaxScaler())])

preproc_pipe=ColumnTransformer([
    ('num_tr', num_transformer, num_features),
    ('cat_tr', cat_transformer, cat_features)])
```

# Pipeline (Baseline Model - Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

pipe=make_pipeline(preproc_pipe,LogisticRegression())
```

```python
#GridSearch the best parameters
from sklearn.model_selection import GridSearchCV

grid={'logisticregression__C':[1, 10, 20, 50],'logisticregression__penalty':["l1","l2","elasticnet"], 'logisticregression__solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}

grid_search = GridSearchCV(pipe, param_grid=grid,cv=5,scoring="accuracy")

grid_search.fit(X, y)
grid_search.best_params_
```

```python
best_pipe=make_pipeline(preproc_pipe,LogisticRegression(C=10,penalty='l1',solver='liblinear'))
```

```python
from sklearn.model_selection import cross_val_score
score_pipe=cross_val_score(best_pipe, X, y, cv=5, scoring='accuracy').mean()
score_pipe
```

```python
best_pipe.fit(X,y)
```

# Predictions

```python
pred=best_pipe.predict(test)
pred
```

```python
df_pred=pd.DataFrame(pred)
df_pred=pd.concat([test['PassengerId'],df_pred],axis=1)
```

```python
df_pred.rename(columns={0: 'Transported'},inplace=True)
df_pred
```

```python
df_pred.to_csv('./data/submission_logisticv2.csv',header=True,index=False)
```
